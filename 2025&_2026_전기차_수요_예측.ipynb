```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# VIF 계산 함수
def calculate_vif(X):
    X = add_constant(X)
    vif = pd.DataFrame()
    vif['Feature'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif.drop(index=0)  # 상수(constant) 제외

# VIF 계산
vif_df = calculate_vif(df.drop(columns=['자치구']))
vif_df.sort_values(by='VIF', ascending=False)
```

#모델링

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor, ExtraTreesRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 독립 변수(X)와 종속 변수(y) 분리
X = data.drop(columns=['전기차(대)'])  # 타겟 열 제외
y = data['전기차(대)']  # 타겟 열

# Train-Test Split (80% 훈련, 20% 테스트)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=99)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 모델 리스트 정의
models = {
    'Gradient Boosting': GradientBoostingRegressor(random_state=123),
    'Random Forest': RandomForestRegressor(random_state=123),
    'Extra Trees': ExtraTreesRegressor(random_state=123)
}

# 변수 중요도 저장 리스트
feature_importances = []

# 모델 훈련 및 평가
results = {}
for name, model in models.items():
    # 모델 훈련
    model.fit(X_train_scaled, y_train)

    # 변수 중요도 저장
    feature_importances.append(model.feature_importances_)

    # 테스트 데이터로 예측
    y_pred = model.predict(X_test_scaled)

    # 평가 지표 계산
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # 결과 저장
    results[name] = {
        'RMSE': rmse,
        'MAE': mae,
        'R2 Score': r2
    }
    print(f"{name} 모델 평가:")
    print(f"  RMSE: {rmse:.2f}")
    print(f"  MAE: {mae:.2f}")
    print(f"  R2 Score: {r2:.2f}\n")

# 결과를 데이터프레임으로 변환하여 보기 좋게 출력
results_df = pd.DataFrame(results).T
print("모델 평가 결과 요약:")
print(results_df)

# 세 모델의 평균 변수 중요도 계산
avg_importance = np.mean(feature_importances, axis=0)
feature_names = X.columns

# 변수 중요도 시각화
plt.figure(figsize=(12, 6))
plt.barh(feature_names, avg_importance, color='skyblue')
plt.xlabel('Average Feature Importance')
plt.ylabel('Features')
plt.title('Average Feature Importance Across Models')
plt.show()
```

# 파라미터 튜닝

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, train_test_split
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import numpy as np
extra_tree = ExtraTreesRegressor(random_state=123)

# 하이퍼파라미터 후보 설정
param_grid = {
    'n_estimators': [100, 300, 500],  # 트리 개수
    'min_samples_leaf': [1, 3, 5],  # 리프 노드의 최소 샘플 수
    'max_features': [1, 3, 6],  # 최대 변수 수
    'max_leaf_nodes': [10, 30, None]  # 최대 리프 노드 수
}

# GridSearchCV를 사용한 하이퍼파라미터 튜닝
grid_search = GridSearchCV(
    estimator=extra_tree,
    param_grid=param_grid,
    cv=10,
    scoring='neg_mean_squared_error',
    verbose=1,
    n_jobs=-1
)
grid_search.fit(X_train_scaled, y_train)

# 최적의 하이퍼파라미터와 성능 확인
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

print("Best Parameters:", best_params)

# 테스트 데이터에서 평가
y_pred = best_model.predict(X_test_scaled)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"RMSE: {rmse:.2f}")
print(f"R2 Score: {r2:.2f}")
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, train_test_split
rf_model = RandomForestRegressor(random_state=123)

# 하이퍼파라미터 후보 설정
param_grid = {
    'n_estimators': [100, 300, 500],  # 트리 개수
    'min_samples_leaf': [1, 3, 5],  # 리프 노드의 최소 샘플 수
    'max_features': [1, 3, 6],  # 최대 변수 수
    'max_leaf_nodes': [10, 30, None]  # 최대 리프 노드 수
}

# GridSearchCV를 사용한 하이퍼파라미터 튜닝
grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    cv=10,
    scoring='neg_mean_squared_error',
    verbose=1,
    n_jobs=-1
)
grid_search.fit(X_train_scaled, y_train)

# 최적의 하이퍼파라미터와 성능 확인
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

print("Best Parameters:", best_params)

# 테스트 데이터에서 평가
y_pred = best_model.predict(X_test_scaled)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"RMSE: {rmse:.2f}")
print(f"R2 Score: {r2:.2f}")
```
