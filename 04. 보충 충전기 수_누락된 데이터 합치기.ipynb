```python
import pandas as pd
```

```python
charger = pd.read_csv('10. 보충 충전기수.csv')
demand = pd.read_csv('predicted_charging_station_demand.csv')
```

```python
charger
```

```python
demand
```

```python
demand = demand.rename(columns = {'급속 충전기 수요': '급속', '완속 충전기 수요': '완속', 
                                  '시설': '대분류', '자치구': '군구'})
demand = pd.melt(demand, id_vars = ['군구', '년도', '대분류'], value_vars = ['급속', '완속'],
                var_name = '기종(대)', value_name = '충전수요')
demand
```

```python
a = demand[(demand['군구'] == '광진구') & (demand['기종(대)'] == '완속') & (demand['대분류']=='마트/쇼핑몰')]
b = demand[(demand['군구'] == '광진구') & (demand['기종(대)'] == '급속') & (demand['대분류']=='마트/쇼핑몰')]
c=demand[(demand['군구'] == '서대문구') & (demand['기종(대)'] == '급속') & (demand['대분류']=='직장')]
d=demand[(demand['군구'] == '동작구') & (demand['기종(대)'] == '급속') & (demand['대분류']=='직장')]
e=demand[(demand['군구'] == '성북구') & (demand['기종(대)'] == '급속') & (demand['대분류']=='직장')]
f=demand[(demand['군구'] == '동대문구') & (demand['기종(대)'] == '급속') & (demand['대분류']=='직장')]
g = demand[(demand['군구'] == '용산구') & (demand['기종(대)'] == '급속') & (demand['대분류']=='마트/쇼핑몰')]
h=demand[(demand['군구'] == '성동구') & (demand['기종(대)'] == '급속') & (demand['대분류']=='마트/쇼핑몰')]
i=demand[(demand['군구'] == '동작구') & (demand['기종(대)'] == '급속') & (demand['대분류']=='마트/쇼핑몰')]
```

```python
df  = pd.concat([a,b,c,d,e,f,g,h,i], axis = 0)
df = df[df['년도']==2026]
df
```

```python
df = df.rename(columns = {'충전수요' : '보충 충전기수'})

df
```

```python
df2 = pd.concat([charger, df])
df2
```

```python
df2.to_csv('보충충전기수_최종.csv')
```
