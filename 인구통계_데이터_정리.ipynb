```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/colab_data/Dscover 24-2 메인프로젝트/서울시_인구통계(2017-2023).csv", encoding = "euc-kr")
df
```

```python
df_melted = pd.melt(df, id_vars=["행정구역별(읍면동)"], var_name="연도_항목", value_name="값")

# 연도와 항목 분리
df_melted[["연도", "항목"]] = df_melted["연도_항목"].str.split(".", expand=True).fillna("0")
df_melted["연도"] = df_melted["연도"].astype(int)

# 피벗 테이블로 정리
df_final = df_melted.pivot(index=["행정구역별(읍면동)", "연도"], columns="항목", values="값").reset_index()
```

```python
df_final
```

```python
new_columns = df_final.iloc[-1, 2:].tolist()  # 마지막 행의 3열부터 열 이름으로 가져옴

df_final.columns = list(df_final.columns[:2]) + new_columns  # 첫 두 열은 기존 열 이름 유지, 나머지는 새 열 이름으로 설정

# "행정구역별(읍면동)" 값이 있는 행 삭제
df_final = df_final[~df_final[df_final.columns[0]].str.contains('행정구역별|서울특별시', na=False)]

# 인덱스 재설정
df_final.reset_index(drop=True, inplace=True)
df_final
```

```python
df_final
```

```python
df_final.iloc[:,1:] = df_final.iloc[:,1:].apply(pd.to_numeric, errors='coerce')
df_final["공동주택"] = (df_final['주택_아파트 (호)'] + df_final["주택_연립주택 (호)"])
df_final["나머지주택"] = (df_final["주택_계 (호)"] - df_final["공동주택"])
df_final["공동주택/가구수"] = df_final["공동주택"] / df_final["가구-계 (가구)"]
df_final["나머지주택/가구수"] = df_final["나머지주택"] / df_final["가구-계 (가구)"]
df_final
```

```python
df_final.info()
```

```python
cols_to_keep = list(df_final.columns[:4]) + list(df_final.columns[-2:])
df_final = df_final[cols_to_keep]
```

```python
df_final.rename(columns={"행정구역별(읍면동)": "자치구"}, inplace=True)
df_final
```
