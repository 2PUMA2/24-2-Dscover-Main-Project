```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/colab_data/Dscover 24-2 메인프로젝트/교통문화지수_시도_시_군_구__20241220152701.csv", encoding = 'euc-kr')
df
```

```python
df_melted = pd.melt(df, id_vars=["행정구역별(2)"], var_name="년도", value_name="값")
df_melted
```
