```python
import pandas as pd
```

```python
data=pd.read_csv('2025,2026 prediction (수정).csv',encoding='euc-kr')
```

```python
data
```

```python
# 일별 평균 전기차 충전소 이용 횟수 정의 (예시)
charging_frequency = {
    '주거지': 0.27,
    '직장': 0.08,
    '마트/쇼핑몰': 0.028,
    '기타시설': 0.16
}

# 시설별 완속/급속 비율 정의
ratio_fast_slow = {
    '주거지': (0.75, 0.25),
    '직장': (0.6, 0.4),
    '마트/쇼핑몰': (0.46, 0.54),
    '기타시설': (0.48, 0.52)
}
```

```python
# 결과를 저장할 리스트 초기화
results = []

# 데이터셋 반복 처리
def calculate_charging_demand(row):
    for facility, freq in charging_frequency.items():
        slow_ratio, fast_ratio = ratio_fast_slow[facility]
        daily_demand = row['전기차(대)'] * freq
        annual_fast_demand = daily_demand * fast_ratio 
        annual_slow_demand = daily_demand * slow_ratio 

        results.append({
            '자치구': row['자치구'],
            '년도': row['년도'],
            '시설': facility,
            '급속 충전기 수요': annual_fast_demand,
            '완속 충전기 수요': annual_slow_demand
        })

# 데이터셋에 대해 계산 적용
data.apply(calculate_charging_demand, axis=1)

# 결과를 DataFrame으로 변환
result_df = pd.DataFrame(results)
```
