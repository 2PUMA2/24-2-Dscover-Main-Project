```python
import pandas as pd
```

```python
df=pd.read_csv('5. 서울시 구별 산업별 사업체수 및 종사자수(2017-2022).csv')
```

```python
df
```

```python
# '자치구별(2)' 행 제거
df = df[df['자치구'] != '자치구별(2)']
```

```python

def fill_future_years(group):
    # 2021, 2022 데이터만 선택
    recent_years = group[group['연도'].isin([2021, 2022])]

    # 숫자형 열의 평균 계산 (NaN 값은 그대로 유지)
    avg_row = recent_years.mean(numeric_only=True).round()

    # 자치구 이름 유지
    avg_row['자치구'] = group['자치구'].iloc[0]

    # 2023과 2024 행 생성
    future_years = [2023, 2024]
    rows = []
    for year in future_years:
        new_row = avg_row.copy()
        new_row['연도'] = year
        rows.append(new_row)

    # 기존 데이터와 새 데이터 결합
    return pd.concat([group, pd.DataFrame(rows)], ignore_index=True)

# 그룹별로 미래 연도 데이터 채우기
df_filled = df.groupby('자치구').apply(fill_future_years).reset_index(drop=True)
```
