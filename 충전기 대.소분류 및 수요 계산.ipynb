```python
# 연식이 8년 미만인 충전기만

df_25 = df[df['2025연식'] < 8]
df_26 = df[df['2026연식'] < 8]
```

```python
result_25 = df_25.groupby(['군구', '기종(대)', '대분류', '이용자제한']).size().unstack(fill_value=0)
result_25
```

```python
result_26 = df_26.groupby(['군구', '기종(대)', '대분류', '이용자제한']).size().unstack(fill_value=0)
result_26
```

```python
result_25.to_csv('2025년 사용가능 대분류별 개수_이용자제한 포함.csv')
result_26.to_csv('2026년 사용가능 대분류별 개수_이용자제한 포함.csv')
```

```python
# 인덱스 리셋
result_25 = result_25.reset_index()
result_26 = result_26.reset_index()

# 년도 컬럼 추가 및 두 데이터 병합
result_25['년도'] = 2025
result_26['년도'] = 2026

result = pd.concat([result_25, result_26], axis = 0)

# 비어있는 컬럼 추가
result['대수'] = np.nan

# 자동차 대수 계산하기 (기타시설은 이용자제한 제외하기)
result = result.reset_index(drop=True)
result.loc[result['대분류'] != '기타시설', '대수'] = result['비공개'] + result['이용가능'] +  result['이용자제한']
result.loc[result['대분류'] == '기타시설', '대수'] = result['비공개'] + result['이용가능'] 
result

# 칼럼정리
result.drop(columns = ['비공개', '이용가능', '이용자제한'], inplace = True)
result
```

```python
# 충전기 수요 연결하기

demand = pd.read_csv('predicted_charging_station_demand.csv')
demand
```

```python
# 컬럼명 정비
demand = demand.rename(columns = {'급속 충전기 수요': '급속', '완속 충전기 수요': '완속', 
                                  '시설': '대분류', '자치구': '군구'})

# 언피벗
demand = pd.melt(demand, id_vars = ['군구', '년도', '대분류'], value_vars = ['급속', '완속'],
                var_name = '기종(대)', value_name = '충전수요')

demand
```

```python
# 수요와 충전기 정보 병합

result = pd.merge(result, demand, on = ['군구', '기종(대)','대분류', '년도'], how = 'left')

# 충전 수요 계산
result['1대당 1일 충전수요'] = result['충전수요'] / result['대수']

result
```

```python
# 완속 급속별 충전 소요 시간

e = ['완속', '급속']
e = pd.Series(e)

f = [ 5, 0.6]
f = pd.Series(f)

time = pd.DataFrame([e,f]).T
time.columns = ['기종(대)', '시간']
time
```

```python
# 충전시간 병합하고 계산하기
result = pd.merge(result, time, on = '기종(대)', how = 'left')

result['1대당 1일 충전시간'] = result['1대당 1일 충전수요'] * result['시간']


result.drop(columns = ['시간'], inplace = True)

result
```

```python
result.to_csv('new_기종별시설별 1대당 충전수요 및 시간(2025, 2026).csv')
```
