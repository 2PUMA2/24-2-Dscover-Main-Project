```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/colab_data/Dscover 24-2 메인프로젝트/사업체현황(산업대분류별+총괄)(2006년+이후)_20241220141601.csv")
df
```

```python
df_melted = pd.melt(df, id_vars=["자치구별(2)", "산업대분류별(1)"], var_name="연도_항목", value_name="값")

# 연도와 항목 분리
df_melted[["연도", "항목"]] = df_melted["연도_항목"].str.split(".", expand=True).fillna("0")
df_melted["연도"] = df_melted["연도"].astype(int)

# 피벗 테이블로 정리
df_final = df_melted.pivot(index=["자치구별(2)", "연도", "산업대분류별(1)"], columns="항목", values="값").reset_index()
```

```python
df_final.columns.name = None
df_final.rename(columns={"자치구별(2)" : "자치구", "산업대분류별(1)": "산업별", "0": "사업체수", "1": "종사자수"}, inplace=True)
df_final
```

```python
# 필요한 데이터만 선택
df_final["사업체수_컬럼"] = df_final["산업별"] + "_사업체수"
df_final["종사자수_컬럼"] = df_final["산업별"] + "_종사자수"
cor_pivot = df_final.pivot(index=["자치구", "연도"], columns="사업체수_컬럼", values="사업체수")
people_pivot = df_final.pivot(index=["자치구", "연도"], columns="종사자수_컬럼", values="종사자수")

# 두 데이터를 합치기
result = pd.concat([cor_pivot, people_pivot], axis=1)

# 열 이름 정리
result.columns.name = None
# 결과 출력

print(result)
```
